```python
import pandas as pd
import numpy as np
data = pd.read_csv('bb1_spatial_indices_big.csv',delimiter = ',',header = 1)

# Listing unique Landsat image IDs from 2013 - 2020
UniqueID_unsorted = np.unique(data['id'])
UniqueID = np.array([1]*(len(UniqueID_unsorted)))
for i in range(len(UniqueID_unsorted)):
    number = int(UniqueID_unsorted[i][12:])
    UniqueID[i] = (number)
UniqueID.sort()

print(f'Available landsat scenes: \n{UniqueID}')
print('\nChoose from these available landsat dates for the first input in sector_plot.py')
```
